```python
import requests
import pandas as pd
import json 

all_players_df = pd.DataFrame()

# Define the list of team IDs
team_ids_list = [2, 40, 25, 29, 1, 28, 6, 32, 5, 15, 7, 37, 3, 20, 8, 30, 4, 27, 9, 11, 34, 17]  # Replace with your actual team IDs
#team_ids_list = [32]
# Initialize an empty DataFrame
all_players_df = pd.DataFrame()

# Define the headers
headers = {
    'accept': 'application/json',
    'User-Agent': 'Your User Agent',  # Set an appropriate user agent
}

# Define the base parameters
base_params = {
    'mode': 'BOTH',
    'records': 40,
    'filterActive': 'true',
    'filterFormatLevel': 'ALL',
    'sort': 'ALPHA_ASC'
}

# Define the URL
url = 'https://hs-consumer-api.espncricinfo.com/v1/pages/player/search'

# Loop through each team ID
for team_id in team_ids_list:
    page = 1
    while True:
        # Update the parameters with the current page and team ID
        params = base_params.copy()
        params.update({
            'page': page,
            'filterTeamId': team_id,
        })

        # Send the request
        response = requests.get(url, verify=False, headers=headers, params=params)
        

        # Parse the response content
        data = response.json()
        
        # Check if 'records' key is in the response data
        if 'results' in data:
            # If there are no more records, break the loop
            if not data['results']:
                break

            
            all_players_df = pd.concat([all_players_df, pd.DataFrame(data['results'])], ignore_index=True)

        # If there are no more records, break the loop
        else:
            print(data)
            break

        
        # Increment the page number
        page += 1
#     
# 
# 
# # Decode the bytes to string and load it as JSON
#     response_json = json.loads(response_content.decode('utf-8'))

#     # Extract the players information
#     players_info = response_json['content']['players']['results']

#     # Convert the list of dictionaries to a DataFrame
#     players_df = pd.DataFrame([{
#         'ID': player['id'],
#         'Name': player['longName'],
#         'Date of Birth': f"{player['dateOfBirth']['year']}-{player['dateOfBirth']['month']}-{player['dateOfBirth']['date']}" if player.get('dateOfBirth') else 'N/A',
#         'Batting Style': ', '.join(player.get('battingStyles', ['N/A'])),
#         'Long Bowling Style': ', '.join(player.get('longBowlingStyles', ['N/A']))
#     } for player in players_info])

#     # Append the current DataFrame to the all_players_df DataFrame
#     all_players_df = pd.concat([all_players_df, players_df], ignore_index=True)

# # Display the combined DataFrame
# print(all_players_df)



#
```
